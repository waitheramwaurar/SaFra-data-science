```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("WEOOct2020all.csv", encoding='cp1252')
df.replace('n/a', 0, inplace=True)
gdp_per_capita_columns = ['Country', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023']
# gdp_per_capita_columns = ['Country', '2013', '2023']
gdp_per_capita_data = df[df['WEO Subject Code'] == 'NGDPPC'][gdp_per_capita_columns]
gdp_per_capita_data
```

```python
for col in gdp_per_capita_columns[1:]:
    gdp_per_capita_data[col] = gdp_per_capita_data[col].replace(',', '', regex=True)
    
gdp_per_capita_data[gdp_per_capita_columns[1:]]   
```

```python
for col in gdp_per_capita_columns[1:]:
    gdp_per_capita_data[col] = pd.to_numeric(gdp_per_capita_data[col])
    
gdp_per_capita_data[gdp_per_capita_columns[1:]]  
```

```python
gdp_per_capita_data
```

```python
#Calculating the percentage growth between the beginning of the decade and the end of the decade
gdp_per_capita_data['Growth Percentage'] = ((gdp_per_capita_data['2023'] - gdp_per_capita_data['2013'])/gdp_per_capita_data['2013'])*100
top10_gdppc_countries = gdp_per_capita_data.sort_values(by='Growth Percentage', ascending=False).head(10)
top10_gdppc_countries[['Country', 'Growth Percentage']]
```

2. Draw OECD countries' "Population" growth over the last decade

```python
oecd_countries = ['Australia', 'Austria', 'Belgium', 'Canada', 'Chile', 'Czech Republic', 'Denmark', 'Estonia', 'Finland', 'France',
                  'Germany', 'Greece', 'Hungary', 'Iceland', 'Ireland', 'Israel', 'Italy', 'Japan', 'Korea', 'Latvia', 'Lithuania',
                  'Luxembourg', 'Mexico', 'Netherlands', 'New Zealand', 'Norway', 'Poland', 'Portugal', 'Slovak Republic', 'Slovenia',
                  'Spain', 'Sweden', 'Switzerland', 'Turkey', 'United Kingdom', 'United States']
pop_growth = df[df['WEO Subject Code'] == 'LP'][gdp_per_capita_columns]
pop_growth.set_index('Country', inplace=True)
pop_growth
```

```python
#Converting the columns format to  numeric
pop_growth = pop_growth.apply(pd.to_numeric, errors='coerce')
pop_growth_pct = pop_growth.pct_change(axis=1) * 100
oecd_pop_growth = pop_growth.loc[oecd_countries]

# plt.figure(figsize(12, 8))
for country in oecd_pop_growth.index:
    plt.plot(oecd_pop_growth.columns, oecd_pop_growth.loc[country], label=country)

plt.title("OECD Countries Population Growth (2014-2023)")
plt.xlabel("Year")
plt.ylabel("Population Growth Rate (%)")
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.show()
```
